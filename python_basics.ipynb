```python
#WAP to print the elements of a list in a single line. (list is the parameter)

cities=['Karachi','Lahore','Islamabad','Quetta','Peshawar']
Islamic_countries=['Kingdom of Saudia Arabia','Pakistan','Qatar','Palestine','Indonesia']
def print_len(list):
    print(len(list))

def print_list(list):
    for item in list:
        print(item , end=" ")

print_len(Islamic_countries)
print_list(Islamic_countries)     
```
